```python
import yfinance as yf

tesla_data = yf.download('TSLA', start='2020-01-01', end='2025-01-01')
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import yfinance as yf
import pandas as pd

tesla = yf.Ticker("TSLA")

# Get quarterly financials (income statement)
tesla_financials = tesla.quarterly_financials

# Extract Revenue and transpose for easy view
tesla_revenue = tesla_financials.loc['Total Revenue'].transpose()
tesla_revenue = tesla_revenue.reset_index()
tesla_revenue.columns = ['Date', 'Revenue']

# Display last 5 rows
tesla_revenue.tail()
```

```python
import yfinance as yf
import pandas as pd

# GameStop ticker
gme = yf.Ticker("GME")

# Get quarterly financials (income statement)
gme_financials = gme.quarterly_financials

# Extract Revenue and transpose for easy view
gme_revenue = gme_financials.loc['Total Revenue'].transpose()
gme_revenue = gme_revenue.reset_index()
gme_revenue.columns = ['Date', 'Revenue']

# Display last 5 rows
gme_revenue.tail()
```
